```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2 * np.pi, 3 * np.pi, 201)

plt.plot(x, np.cos(x - 5), 'b')
plt.plot(x, 3 * np.cos(x - 2) + 3, 'y')
plt.plot(x, 2 * np.cos(x) - 6, 'g')

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True) 
```

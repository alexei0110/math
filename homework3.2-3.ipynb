```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pol_to_dec(r, phi):
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return (x, y)

print(pol_to_dec(20, 45))
```

```python
phi = np.arange(0., 2., 1./180.)*np.pi
plt.polar(phi, [10] * len(phi))
plt.show()
```

```python
x = np.linspace(0, 5, 100)
y = 3 * x + 2
plt.polar(x, y)
plt.show()
```

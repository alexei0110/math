```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.plot(x, x ** 2 - 1)
         
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

```python
from scipy.optimize import fsolve
def equations(p):
    x, y = p
    return (y - x ** 2 + 1, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (1, 1))
x2, y2 = fsolve(equations, (-2, 2))

print(f'(x1, y1) = ({x1}, {y1})')
print(f'(x2, y2) = ({x2}, {y2})')
```

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (-1 / x) + 2)
plt.plot(x, x ** 2 - 1)
         
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 5)
plt.grid(True)

plt.show()
```

```python
from scipy.optimize import fsolve
def equations(p):
    x, y = p
    return (y - x ** 2 + 1, (-1 / x) + 2 - y)

x1, y1 = fsolve(equations, (2, 2))
x2, y2 = fsolve(equations, (0.5, -0.5))
x3, y3 = fsolve(equations, (-2, 2))

print(f'(x1, y1) = ({x1}, {y1})')
print(f'(x2, y2) = ({x2}, {y2})')
print(f'(x3, y3) = ({x3}, {y3})')
```
